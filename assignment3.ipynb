```python
#preprocessing raw data
file=open("S:\\INFORMATION TECHNOLOGY\\3rd sem\\machine learning in biological processes\\assignment2\\pr_data.txt","r")
raw_data=file.read()
raw_data=list(raw_data)
i=0
while(i<len(raw_data)):
    if((ord(raw_data[i])>=ord('0') and ord(raw_data[i])<=ord('9')) or raw_data[i]=="\n" or raw_data[i]=="'" or raw_data[i]==' ' or raw_data[i]=="'"):
        raw_data.pop(i)
        continue
    if(raw_data[i]=='?' or raw_data[i]=='_'):
        raw_data[i]='U'
    if(raw_data[i]=='S' and raw_data[i+1]=='t' and raw_data[i+2]=='r'):
        raw_data[i+1]='e'
        raw_data[i+2]='q'
        i+=2
    i+=1
raw_data="".join(raw_data)
raw_data=raw_data.split("Seq{}=")
raw_data.pop(0)
Seq=[]
SS=[]
for i in range(len(raw_data)):
    if(i%2==0):
        Seq.append(raw_data[i])
    else:
        SS.append(raw_data[i])
file.close()
# print(Seq)
print(SS)
```

```python
#Adding dummy
window_size=5
dummy_Seq=Seq
pad_len=(window_size-1)//2
for i in range(len(dummy_Seq)):
    dummy_Seq[i]="X"*pad_len+dummy_Seq[i]+"X"*pad_len
```

```python
#Training and Testing Data
Data_in=[]
Data_out=[]
j=0
for i in dummy_Seq:
    for k in range(len(SS[j])):
        Data_in.append(i[k:k+window_size])
        Data_out.append(SS[j][k])
    j+=1
len_train_data=int(len(Data_in)*0.7)
Training_Data_in=Data_in[:len_train_data]
Training_Data_out=Data_out[:len_train_data]
Test_Data_in=Data_in[len_train_data:]
Test_Data_out=Data_out[len_train_data:]
print("input training data:","\t","output training data:","\t","input testing data:","\t","output testing data:")
for i in range(len(Training_Data_in)):
    if(i<len(Test_Data_in)):
        print(Training_Data_in[i],"\t","\t","\t",Training_Data_out[i],"\t","\t","\t",Test_Data_in[i],"\t","\t","\t",Test_Data_out[i])
    else:
        print(Training_Data_in[i],"\t","\t","\t",Training_Data_out[i])
```

```python
#Class Wise Data
SS_list=['G','H','I','B','E','T','S','U']
SS_dict={}
for i in SS_list:
    SS_dict[i]=[]
for i in range(len(Training_Data_out)):
    SS_dict[Training_Data_out[i]].append(Training_Data_in[i])
# print(SS_dict['I'])
    
```

```python
#20*L size probability Table
amino_acids="ARNDCQEGHILKMFPSTWYVX"
amino_acids_count={}
for i in SS_list:
    amino_acids_count[i]={}
    for j in amino_acids:
        amino_acids_count[i][j]=[0 for i in range(window_size)]
class_probability={}
Probability_tables={}
for i in SS_list:
    Probability_tables[i]={}
    class_probability[i]=len(SS_dict[i])/len(Training_Data_in)
# print(class_probability)
for i in SS_list:
    for j in amino_acids:
        Probability_tables[i][j]=[0 for i in range(window_size)]
# l=[0 for i in range(window_size)]
for i in SS_list:
    for j in SS_dict[i]:
        for k in range(window_size):
            amino_acids_count[i][j[k]][k]+=1
            # l[k]+=1
# print(l)
# print(amino_acids_count)
for i in SS_list:
    for j in amino_acids:
        for k in range(window_size):
            if(len(SS_dict[i])>0):
                Probability_tables[i][j][k]=amino_acids_count[i][j][k]/len(SS_dict[i])
            else:
                Probability_tables[i][j][k]=0
print(Probability_tables)
```

```python
#Testing the test data and finding Accuracy
Test_data_class=[[0 for i in range(len(SS_list))] for j in range(len(Test_Data_in))]
for i in range(len(Test_Data_in)):
    for j in range(len(SS_list)):
        sum=1
        for k in range(len(Test_Data_in[i])):
            sum*=Probability_tables[SS_list[j]][Test_Data_in[i][k]][k]
        sum*=class_probability[SS_list[j]]
        Test_data_class[i][j]=sum
print(SS_list)
count=0
print("Sequence","Original","Calculated")
for i in range(len(Test_Data_in)):
    print(Test_Data_in[i],"\t ",Test_Data_out[i],"\t ",SS_list[Test_data_class[i].index(max(Test_data_class[i]))])
    if(Test_Data_out[i]==SS_list[Test_data_class[i].index(max(Test_data_class[i]))]):
        count+=1
print("Accuracy is: ",count/len(Test_Data_in))

```
